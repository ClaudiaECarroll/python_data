# Data Cleaning

```python
import pandas as pd
```

```python
df_SAFI = pd.read_csv("/content/drive/MyDrive/workshop_data/SAFI_results.csv")
```

```python
df_SAFI.head()
```

```python
df_SAFI.columns.values
```

```python
#Removing codes from start of column headers

df_SAFI.columns = df_SAFI.columns.str.replace(r'^.*?_', '', regex=True)
```

```python
df_SAFI.columns.values
```

```python
#Checking column headers

df_SAFI.info()
```

```python
df_SAFI = df_SAFI[['years_farm',
       'respondent_roof_type', 'respondent_wall_type',
       'respondent_floor_type', 'buildings_in_compound',
       'rooms', 'other_buildings',
        'plots_count', 'note',
        'memb_assoc']]
```

```python
df_SAFI.head()
```

```python
df_SAFI.info()
```

```python
df_SAFI['memb_assoc']
```

```python
#Remove rows with a NaN value in a particular column

df_SAFI = df_SAFI[df_SAFI['memb_assoc'].notna()]

df_SAFI.info()
```

```python
#Use dropna() to remove all columns with NaN values

df_SAFI.dropna(inplace=True, axis="columns")
df_SAFI.info()
```

```python
# Save cleaned dataframe to a new csv file

df_SAFI.to_csv("/content/drive/MyDrive/workshop_data/SAFI_results_cleaned.csv")
```

```python
#Using describe to count how many cells in a column have non-null values

df_SAFI['years_farm'].describe()
```
